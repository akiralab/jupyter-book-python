```python
dice_probabilities = np.array([[sums.count(i+j)/36 for j in range(1, 7)] for i in range(1, 7)]) * 100

# 確率ベースのヒートマップで可視化、x ticksを上に表示し、ラベルを1~6に変更
plt.figure(figsize=(8,6))
ax = sns.heatmap(dice_probabilities, annot=True, fmt=".1f", cmap="YlGnBu", square=True, cbar=True)
ax.set_title('サイコロの目の和の確率のヒートマップ')
ax.set_xlabel('1つ目のサイコロ')
ax.set_ylabel('2つ目のサイコロ')
ax.set_xticklabels(range(1, 7))  # X軸のラベルを1~6に変更
ax.set_yticklabels(range(1, 7))  # Y軸のラベルを1~6に変更
ax.xaxis.tick_top()  # x ticksを上側に表示
ax.xaxis.set_label_position('top')  # x軸のラベルを上に移動
plt.show()
```

```python
# 各セルに「和（確率%）」の形式で表示するためのデータを作成
formatted_probabilities = np.array([[f"{i+j} ({sums.count(i+j)/36*100:.0f}%)" for j in range(1, 7)] for i in range(1, 7)])

# フォーマットを適用してヒートマップで可視化
plt.figure(figsize=(10,8))
ax = sns.heatmap(dice_probabilities, annot=formatted_probabilities, fmt="", cmap="YlGnBu", square=True, cbar=True)
ax.set_title('サイコロの目の和の確率のヒートマップ')
ax.set_xlabel('サイコロ2')
ax.set_ylabel('サイコロ1')
ax.set_xticklabels(range(1, 7))  # X軸のラベルを1~6に変更
ax.set_yticklabels(range(1, 7))  # Y軸のラベルを1~6に変更
ax.xaxis.tick_top()  # x ticksを上側に表示
ax.xaxis.set_label_position('top')  # x軸のラベルを上に移動
plt.show()
```

オッズ比は、特定の和が出る確率が2が出る確率に対してどれだけ多いかを示します。例えば、和が7の場合のオッズ比は6.00となり、これは2が出る確率の6倍であることを意味します。
