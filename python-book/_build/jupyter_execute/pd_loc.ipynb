# loc

```python
import pandas as pd
import os
```

```python
# pathの確認
path = '../src/titanic.csv'
os.path.exists(path)
```

### read_csv()

```python
# 氏名をindexとして指定してみる
df = pd.read_csv(path, index_col=3)
df.head()
```

### loc["item名"] -> itemについての情報を取り出すことができる

```python
df.loc['Heikkinen, Miss. Laina']
```

### データを絞り込むことも可能

```python
df_survived = df.loc[df["Survived"] == 1]
df_survived
```

```python
df_survived_female = df.loc[(df["Survived"] == 1) & (df["Sex"] == "female")]
df_survived_female
```
